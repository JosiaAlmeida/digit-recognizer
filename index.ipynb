```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
base = pd.read_csv('train.csv')
base.head()
```

```python
prevision = base.iloc[:, 1:].values
classe = base.iloc[:, 0].values
classe
```

```python
prevision
```

```python
x_train, x_test, y_train, y_test = train_test_split(prevision, classe, test_size=0.2, random_state=42)
x_train
```

```python
modelo = svm.SVC()
modelo.fit(x_train, y_train)
modelo
```

```python
prevision = modelo.predict(x_test)
```

```python
score = r2_score(y_test, prevision)
errorScore = mean_squared_error(y_test, prevision)
print(score, errorScore)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
score = accuracy_score(y_test, prevision)
scoreerror = 1- score

print(score, scoreerror)
```
